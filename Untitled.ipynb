```python
modes = ["mode1", "mode2", "mode3"]
conditions = [
    (lambda m: False),
    (lambda m: m == "mode2")
]
kept = []
for mode in modes:
    for condition_fn in conditions:
        if condition_fn(mode):
            break
    else:
        kept.append(mode)

print(kept)
```
